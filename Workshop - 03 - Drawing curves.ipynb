```python
from math import cos

y=list()
x=list()
for i in range(500):
    x.append(i)
    y.append(cos(i/10.0))
```

```python
y[1:10]
```

```python
plt.plot(x, y)
plt.show()
```

```python
from math import exp
y1=list()
y2=list()
x=list()
for i in range(500):
    x.append(i)
    carrier = cos(i/10.0)
    signal = exp(-abs(i-250)/100.0)
    y1.append(signal)
    y2.append(carrier*signal)
plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```

```python
plt.plot(x, y1 , label = "signal")
plt.plot(x, y2, label = "final")
plt.legend(loc='upper right')
plt.show()
```
